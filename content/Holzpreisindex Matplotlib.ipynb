```python
# this workaround is needed because of
# https://github.com/jupyterlite/jupyterlite/issues/119

from js import fetch

URL = "https://raw.githubusercontent.com/ronnystandtke/jupyter-exercises/main/content/Holzpreisindex.csv"
res = await fetch(URL)
text = await res.text()

filename = 'data.csv'

with open(filename, 'w') as f:
    f.write(text)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df = pd.read_csv(filename)

fig, ax = plt.subplots(figsize=(17, 7))
fig.autofmt_xdate()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))

plt.plot(df.Periode, df["Baumaterialien Holz"],   color="blue",  linewidth=2, linestyle="-")
plt.plot(df.Periode, df["Hilfsmaterialien Holz"], color="green", linewidth=1, linestyle=":")
plt.plot(df.Periode, df["Verpackungen Holz"],     color="red",   linewidth=1, linestyle="-.")

display(df.head(5))
plt.show()
```
